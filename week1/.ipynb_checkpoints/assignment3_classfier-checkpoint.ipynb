```python
import pandas as pd
import openpyxl
# Excel 파일 불러오기
df = pd.read_excel('C:/Users/seongmun/AiStudy/AI_Study/data/Raisin_Dataset.xlsx', index_col=0)
df
```

```python
y = df['Class']
y
```

```python
x = df
x.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
# x에는 입력데이터, y에는 타깃 데이터(모델이 예측하려는 실제 정답값)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
# x_train : 훈련 데이터의 입력 특성
# y_train : 훈련 데이터의 정답(타깃 데이터 or 레이블)
# x_test : 예측 데이터의 입력 특성
# y_test : 예측 데이터의 정답(타깃 데이터 or 레이블)
# x : (입력 데이터 갯수, feature 갯수)
# y : (타깃 데이터 갯수)
x_train_encoded = pd.get_dummies(x_train)
x_test_encoded = pd.get_dummies(x_test)
# 문자열 인코딩
# 정답 타깃인 y는 이코딩 안해도 됨. -> 입력 특성 데이터에만 적용
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
# DT
from sklearn.tree import DecisionTreeClassifier

clf_dt = DecisionTreeClassifier(random_state=0)
clf_dt.fit(x_train_encoded, y_train)

pred_dt = clf_dt.predict(x_test_encoded)

print ("\n--- Decision Tree Classifier ---")
print (accuracy_score(y_test, pred_dt))
print (confusion_matrix(y_test, pred_dt))
```

```python
# RT
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

print ("\n--- Random Forest ---")
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(x_train_encoded, y_train)
pred = rf_clf.predict(x_test_encoded)
print(accuracy_score(y_test,pred))
print (confusion_matrix(y_test, pred))
```

```python
# LR
from sklearn.linear_model import LogisticRegression

clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(x_train_encoded, y_train)

pred_lr = clf_lr.predict(x_test_encoded)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```

```python
# SVM
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC

clf_svm = SVC(random_state=0)
clf_svm.fit(x_train_encoded, y_train)

pred_svm = clf_svm.predict(x_test_encoded)

print("\n--- SVM Classifier ---")
print(accuracy_score(y_test, pred_svm))
print(confusion_matrix(y_test, pred_svm))
```
